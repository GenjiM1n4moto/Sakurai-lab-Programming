```python
last_score = int(input('Your last score:\n'))
score_this_time = int(input('Your score this time:\n'))
score_gap = score_this_time - last_score
print('The score gap between the two tests;\n'+ str(score_gap))
```

# 6. AttributeError: 'tuple' object has no attribute 'remove'

```python
user = (0,1,2,3,4,5)
user.remove(1)
print(user)
```

Although tuples and lists belong to sequences, because tuples are immutable sequences, the elements in tuples cannot be modified unless they are replaced as a whole.  
Lists can use append(), extend(), insert(), remove(), and pop() methods to modify elements, but tuples cannot.
